In this experiment, mean, median, and Gaussian filters were applied to two digital images in order to reduce noise and analyze their performance. All three filters were implemented using a **5 Ã— 5 kernel**. The mean filter smooths the image by averaging neighboring pixel values, the median filter replaces each pixel with the median value of its neighborhood to reduce impulse noise, and the Gaussian filter performs weighted smoothing based on a Gaussian distribution. The given images contain a **high level of random noise**, which limits the visible improvement when a small kernel size is used. However, among the applied filters, the **Gaussian filter** produces smoother and more visually balanced results compared to the mean and median filters.

```python
# reading images
img1 = cv2.imread('image1.jpg')
img2 = cv2.imread('image2.jpg')

# Convert BGR to RGB for proper display
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(img1)
plt.title("Original Image 1")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img2)
plt.title("Original Image 2")
plt.axis('off')

plt.show()
```

```python
# apply filters

# Mean Filter
mean1 = cv2.blur(img1, (5,5))
mean2 = cv2.blur(img2, (5,5))

# Median Filter
median1 = cv2.medianBlur(img1, 5)
median2 = cv2.medianBlur(img2, 5)

# Gaussian Filter
gaussian1 = cv2.GaussianBlur(img1, (5,5), 0)
gaussian2 = cv2.GaussianBlur(img2, (5,5), 0)
```

```python
# results display

def show_results(original, mean, median, gaussian, title):
    plt.figure(figsize=(15,5))

    plt.subplot(1,4,1)
    plt.imshow(original)
    plt.title('Original')
    plt.axis('off')

    plt.subplot(1,4,2)
    plt.imshow(mean)
    plt.title('Mean Filter')
    plt.axis('off')

    plt.subplot(1,4,3)
    plt.imshow(median)
    plt.title('Median Filter')
    plt.axis('off')

    plt.subplot(1,4,4)
    plt.imshow(gaussian)
    plt.title('Gaussian Filter')
    plt.axis('off')

    plt.suptitle(title, fontsize=16)
    plt.show()
```

```python
show_results(img1, mean1, median1, gaussian1, "Image 1 Filtering Results")
show_results(img2, mean2, median2, gaussian2, "Image 2 Filtering Results")
```
